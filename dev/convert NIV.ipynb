## convert NIV format

Don't run script in this file. NIV.txt has been already converted. 

```python
f = open('../py_kbible/data/NIV.txt')
lines = f.readlines()
```

```python
lines[0:10]
```

```python
def tidy(lines, start=0, stop=-1, filename='NIV.txt'):
    book = ''
    chapter = 1
    verse = 1
    text = ''
    f = open(filename, 'w')

    for l in lines[start:stop]:
        if l.find('[') > -1:
            tmp = l[1:l.find(']')]
            book = tmp[:tmp.rfind(' ')]
            chapter = int(tmp[tmp.rfind(' ')+1:])
            continue
        dotp = l.find('.')
        if (dotp > -1) and (dotp < 10):
            verse = int(l[:dotp])
            text = l.strip('\n\r')[dotp+1:]

            msg = '{}{}:{} {}\n'.format(book, chapter, verse, text)
            f.write(msg)
            #print('{}'.format(l))
```

```python
tidy(lines)
```
